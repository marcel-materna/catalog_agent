```python
import duckdb
```

```python
# Verbindung zur bestehenden DuckDB-Datei herstellen
con = duckdb.connect(r"C:\Projekte\agent_duckdb\catalog_agent\data\metadata.duckdb")

# Tabellen anzeigen
print(con.execute("SHOW TABLES").fetchall())

# Daten aus einer Tabelle lesen
df = con.execute("SELECT * FROM energy_metadata WHERE TRIM(data_category) = 'Ladedaten'").fetchdf()

# Optional: Ersten Datensatz anzeigen
display(df.head())

# Verbindung schließen
con.close()
```
